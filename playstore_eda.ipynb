# **3. Data Wrangling**

```python
#writing the code to make the dataset analysis ready
#creating one function that can evalute Price,Installs,Size and Reviews
def convert_int(x):
  '''this function takes the object and removes ',',+,M,k,$ returns it after evaluting
  by the function'''
  if x.isdigit():
    return eval(x)
  else:
    if ',' in x:
      x=x.replace(',','')
    if '+' in x:
      x=x.replace('+','')
    if 'M' in x:
      x=x.replace('M','')
      x=eval(x)
      x=x * 10**6
      return x
    if 'k' in x:
      x=x.replace('k','')
      x=eval(x)
      x=x * 10*3 
      return x
    if '$' in x:
      x=x.replace('$','')
    if x=='Free' or x=='Varies with device' or x=='Everyone':
      x=0
    else:
      x=eval(x)
    return x      
```

```python
df_1['Reviews']=df_1['Reviews'].apply(convert_int)
df_1['Size']=df_1['Size'].apply(convert_int)
df_1['Installs']=df_1['Installs'].apply(convert_int)
df_1['Price']=df_1['Price'].apply(convert_int)
```
